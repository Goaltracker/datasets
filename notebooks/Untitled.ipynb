```python
import ddf_utils
import numpy as np
import pandas as pd

output = "ddf-goaltracker-sweden/"

df = pd.read_csv("sweden/data.csv")

#pd.isnull(df.var2)df["Cou#ntry"][2]
```

```python
groupedIndicator = df.groupby(" SeriesCode")

for i, g in groupedIndicator:
     
    measure = i.lower()
    
   # if i == "AG_FPA_CFPI":
    #print(g)
    g[" GeoAreaName"] = "swe"
    
    #print(g["Country"])
    g = g[pd.isnull(g["Country"])]
   
    #print(g[" Value"].size)
    
    fixedG = g[[" GeoAreaName", " TimePeriod", " Value"]]
    fixedG.columns = ["geo", "year", measure]
    
    print(fixedG[measure].size)
   
    if fixedG[measure].size != 0:
    #print(fixedG)

        fixedG.to_csv(output + "ddf--datapoints--" + i.lower() + "--by--geo--year.csv", index=False)


        
```
